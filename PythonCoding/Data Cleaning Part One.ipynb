1. CALCULATING PERCENTAGE OF MISSING VALUES IN EACH COLUMN 

```python
count = data.isnull().sum()
percent = round(count/170350*100 , 2)
k = [count , percent]
missing_count = pd.concat(k, axis=1, keys=['Missing_Values','Percent'])
missing_count.sort_values(by='Missing_Values', ascending=False)
```

Many columns contain huge percentage of missing data <br>
Not efficient to handle columns with majority of the entries missing <br>

2 . REMOVING COLUMNS WITH MAJORITY OF DATA MISSING

```python
# Function to remove columns with MORE THAN 50% MISSING VALUES
def drop_missing_columns(cols):
    if data[cols].isnull().sum() > data[cols].count() / 2 :
        del data[cols]
```

```python
# Using function for each column in the dataset
for i in data.columns:
    drop_missing_columns(i)
```
